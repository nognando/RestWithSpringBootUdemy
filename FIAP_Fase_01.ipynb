```python
#AULA 1 - Aquisição e Leitura de Dados Reais
import pandas as pd #1ª célula

#dados = pd.read_csv('/content/A150850189_28_143_208.csv', encoding='ISO-8859-1', skiprows=3, sep=';', skipfooter=12, thousands='.', decimal=',') #2ª célula
dados = pd.read_csv('/content/A152918189_28_143_208.csv', encoding='ISO-8859-1', skiprows=3, sep=';', skipfooter=12, thousands='.', decimal=',', engine='python') #2ª célula

#Formantando o arquivo com casas decimais
pd.options.display.float_format = '{:.2f}'.format

#lendo oarquivo
dados.head()#3ª célula

```

```python
# traz os 5 ultimos registros
dados.tail()
```

```python
dados.info()
```

```python
# Seleciona todas as colunas do DataFrame, exceto a primeira (índice 0)
colunas_para_converter = dados.columns[1:]

# Converte todas as outras colunas para tipo numérico
# Qualquer valor que não puder ser convertido será transformado em NaN (erro ignorado)
dados[colunas_para_converter] = dados[colunas_para_converter].apply(pd.to_numeric, errors='coerce')
```

```python
# Calcula a média de todas as colunas, exceto a primeira (índice 0)
# Isso ignora, por exemplo, a coluna com o nome dos estados
media = dados.iloc[:, 1:].mean()

# Define um nome para a Series resultante da média, útil para visualização
media.name = 'Média Geral'

# Exibe a Series com as médias e o nome definido
display(media)
```

```python
# Converte os nomes das colunas para datas (pandas tenta interpretar)
# Seleciona os nomes das colunas, exceto a primeira (índice 0)
colunas_data = dados.columns[1:]
datas_convertidas = pd.to_datetime(colunas_data, format="%Y/%b", errors='coerce')

# Encontra a posição da data mais recente
indice_max = datas_convertidas.argmax()

# Pega o nome da coluna correspondente à data mais recente
coluna_mais_recente = colunas_data[indice_max]

print(f"Coluna mais recente: {coluna_mais_recente}")
```
